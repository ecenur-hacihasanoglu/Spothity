```python
# ============================================================
# 0) GEREKLİ KÜTÜPHANELER
# ============================================================
import pandas as pd, numpy as np
from scipy import stats
from statsmodels.tsa.stattools import adfuller

# ============================================================
# 1) SAYISAL SÜTUN LİSTESİ  
# ============================================================
numeric_cols = [
    "duration_ms", "artist_followers",
    "af_danceability","af_energy","af_loudness",
    "af_speechiness","af_acousticness","af_instrumentalness",
    "af_liveness","af_valence","af_tempo",
    "bert_sent_score","positivity","negativity","emotional_intensity",
    "bert_admiration","bert_anger","bert_excitement","bert_fear",
    "bert_joy","bert_love","bert_neutral","bert_sadness"
]
num_features = numeric_cols + ["available_markets_len"]

# ----- available_markets uzunluğunu ekle -----
df_proc = df.copy()
df_proc["available_markets_len"] = (
    df_proc["available_markets"]
          .fillna("[]")
          .apply(lambda s: len(eval(s)) if isinstance(s, str) else len(s))
)

# ============================================================
# 2)  TEK SÜTUN ANALİZ FONKSİYONU  (LogNorm için güvenli)
# ============================================================
def analyse_series(s: pd.Series) -> pd.Series:
    s = s.dropna()
    out = dict(
        mean       = s.mean(),
        std        = s.std(),
        skew       = s.skew(),
        kurtosis   = s.kurtosis()
    )

    # ── Normalite ────────────────────────────────
    out["sw_p"]        = stats.shapiro(s)[1]      if len(s) < 5000 else np.nan
    out["dagostino_p"] = stats.normaltest(s)[1]
    out["ks_norm_p"]   = stats.kstest(
        s, "norm", args=(s.mean(), s.std(ddof=1))
    )[1]

    # ── Poisson ──────────────────────────────────
    lam = s.mean()
    out["lam"]        = lam
    out["var"]        = s.var()
    out["ks_pois_p"]  = stats.kstest(s, "poisson", args=(lam,))[1]

    # ── Durağanlık (ADF) ─────────────────────────
    try:
        out["adf_p"] = adfuller(s, autolag="AIC")[1]
    except ValueError:                        # seri çok kısa vs.
        out["adf_p"] = np.nan

    # ── Çoklu aday dağılımlar (KS D istatistiği) ─
    cand = {
        "Normal" : stats.kstest(
            s, "norm", args=(s.mean(), s.std(ddof=1))
        )[0],
        "Poisson": stats.kstest(s, "poisson", args=(lam,))[0],
        "Exp"    : stats.kstest(s, "expon",  args=stats.expon.fit(s))[0],
    }

    # Log-Normal yalnızca s>0 ise dene
    if (s > 0).all():
        try:
            lognorm_params   = stats.lognorm.fit(s, floc=0)
            cand["LogNorm"]  = stats.kstest(
                s, "lognorm", args=lognorm_params
            )[0]
        except Exception:
            cand["LogNorm"] = np.inf
    else:
        cand["LogNorm"] = np.inf

    out["best_dist"] = min(cand, key=cand.get)
    return pd.Series(out)


# sayısal sütunlar için tablo

dist_summary = df_proc[num_features].apply(analyse_series).T
dist_summary


```

```python
import pandas as pd, numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
from lightgbm import LGBMRegressor
import joblib, pathlib, sklearn
from packaging import version
from scipy import sparse
# Kullanılacak sütunlar
USE_COLS = [
    "date",
    "duration_ms", 
    "explicit", 
    "available_markets",
    "af_danceability", "af_energy", "af_key", "af_loudness", "af_mode","af_speechiness", "af_acousticness", "af_instrumentalness","af_liveness", "af_valence", "af_tempo", "af_time_signature",
    "artist_followers",
    "title_topic", "lyrics_topic",
    "bert_sent_label",
    "bert_sent_score", "positivity", "negativity", "emotional_intensity","bert_admiration", "bert_anger", "bert_excitement", "bert_fear","bert_joy", "bert_love", "bert_neutral", "bert_sadness"
]
TARGET = "popularity"

df = df[USE_COLS + [TARGET]].copy()

# Sütun grupları
numeric_cols = [
    "duration_ms", "artist_followers",
    "af_danceability","af_energy","af_loudness",
    "af_speechiness","af_acousticness","af_instrumentalness",
    "af_liveness","af_valence","af_tempo",
    "bert_sent_score","positivity","negativity","emotional_intensity",
    "bert_admiration","bert_anger","bert_excitement","bert_fear",
    "bert_joy","bert_love","bert_neutral","bert_sadness"
]

low_card_cat = [
    "explicit","af_key","af_mode","af_time_signature",
    "title_topic","lyrics_topic","bert_sent_label"
]

list_cols = ["available_markets"]


# Dönüştürücü: available_markets -> uzunluk
# ListLength transformer
class ListLength(BaseEstimator, TransformerMixin):
    def __init__(self, col="available_markets"): self.col = col
    def fit(self, X, y=None): return self
    def transform(self, X):
        out = X.copy()
        out[self.col] = out[self.col].fillna("[]")                 \
                        .apply(lambda s: len(eval(s)) if isinstance(s,str) else len(s))
        return out

# Tarih → datetime  +  split indexleri

import pandas as pd, numpy as np, joblib, pathlib, sklearn
from packaging import version
from sklearn.pipeline      import Pipeline
from sklearn.compose       import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute        import SimpleImputer
from sklearn.base          import BaseEstimator, TransformerMixin
from scipy import sparse

# ensure date dtype
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# chrono split indices
train_idx = df["date"].dt.year.isin([2017, 2018])
val_idx   = df["date"].dt.year == 2019
test_idx  = df["date"].dt.year == 2020

X_train_raw = df.loc[train_idx, USE_COLS]
y_train     = df.loc[train_idx, TARGET].values
X_val_raw   = df.loc[val_idx,   USE_COLS]
y_val       = df.loc[val_idx,   TARGET].values
X_test_raw  = df.loc[test_idx,  USE_COLS]
y_test      = df.loc[test_idx,  TARGET].values

# custom transformer for available_markets length
class ListLength(BaseEstimator, TransformerMixin):
    def __init__(self, col="available_markets"): self.col = col
    def fit(self, X, y=None): return self
    def transform(self, X):
        out = X.copy()
        out[self.col] = (
            out[self.col].fillna("[]")
                         .apply(lambda s: len(eval(s)) if isinstance(s, str) else len(s))
        )
        return out

# pipelines
if version.parse(sklearn.__version__) >= version.parse("1.4"):
    ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
else:
    ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)

numeric_pipe = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("sc",  StandardScaler())
])
categorical_pipe = Pipeline([
    ("imp", SimpleImputer(strategy="most_frequent")),
    ("ohe", ohe)
])
listlen_pipe = Pipeline([
    ("len", ListLength()),
    ("imp", SimpleImputer(strategy="median")),
    ("sc",  StandardScaler())
])

preproc = ColumnTransformer([
    ("num",     numeric_pipe,   numeric_cols),
    ("cat",     categorical_pipe, low_card_cat),
    ("markets", listlen_pipe,   ["available_markets"])
], sparse_threshold=0.0)        # force dense

# fit / transform
X_train = preproc.fit_transform(X_train_raw)
X_val   = preproc.transform(X_val_raw)
X_test  = preproc.transform(X_test_raw)
if sparse.issparse(X_train):
    X_train, X_val, X_test = map(np.asarray, (X_train, X_val, X_test))

# save
pathlib.Path("prepared_data").mkdir(exist_ok=True)
np.savez_compressed("prepared_data/sets.npz",
                    X_train=X_train, y_train=y_train,
                    X_val=X_val,     y_val=y_val,
                    X_test=X_test,   y_test=y_test)
joblib.dump(preproc, "prepared_data/preprocessor.joblib")
print(" prepared_data/sets.npz  +  preprocessor.joblib written")
# -------------------------------------------------------------------------
```

```python

import numpy as np, joblib
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# load prepared data
data = np.load("prepared_data/sets.npz")
X_train, y_train = data["X_train"], data["y_train"]
X_val,   y_val   = data["X_val"],   data["y_val"]
X_test,  y_test  = data["X_test"],  data["y_test"]

# model
lgbm = LGBMRegressor(n_estimators=50, learning_rate=0.05,
                     num_leaves=24, subsample=0.9,
                     colsample_bytree=0.8, random_state=42)
lgbm.fit(X_train, y_train)

# metrics helper
rmse = lambda yt, yp: np.sqrt(mean_squared_error(yt, yp))

# evaluate
val_pred  = lgbm.predict(X_val)
test_pred = lgbm.predict(X_test)

print("\n=== Validation 2019 ===")
print(f"MAE : {mean_absolute_error(y_val, val_pred):.2f}")
print(f"RMSE: {rmse(y_val, val_pred):.2f}")
print(f"R²  : {r2_score(y_val, val_pred):.3f}")

print("\n=== Test 2020 ===")
print(f"MAE : {mean_absolute_error(y_test, test_pred):.2f}")
print(f"RMSE: {rmse(y_test, test_pred):.2f}")
print(f"R²  : {r2_score(y_test, test_pred):.3f}")
# -------------------------------------------------------------------------
```

```python
# ------------  DNN CELL  -----------------------------------------------
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1)  LOAD PREPARED DATA  (dense NumPy arrays)
data    = np.load("prepared_data/sets.npz")

top20_cols = imp_df["feature"].tolist()       

# İsim -> indeks haritası oluştur
idx_map   = {name: i for i, name in enumerate(feature_names)}



X_train, y_train = data["X_train"][:, :], data["y_train"]
X_val,   y_val   = data["X_val"][:,   :], data["y_val"]
X_test,  y_test  = data["X_test"][:,  :], data["y_test"]


print(f"Loaded shapes  •  Train {X_train.shape}  Val {X_val.shape}  Test {X_test.shape}")

# 2)  BUILD MLP
n_features = X_train.shape[1]
inp = keras.Input(shape=(n_features,))
x   = layers.Dense(256, activation="relu")(inp)
x   = layers.BatchNormalization()(x)
x   = layers.Dropout(0.2)(x)
x   = layers.Dense(128, activation="relu")(x)
x   = layers.BatchNormalization()(x)
x   = layers.Dropout(0.2)(x)
x   = layers.Dense(64, activation="relu")(x)
x   = layers.BatchNormalization()(x)
out = layers.Dense(1)(x)

model = keras.Model(inp, out)
model.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss="mae",
              metrics=[keras.metrics.RootMeanSquaredError(name="rmse")])
model.summary(line_length=110)

# 3)  TRAIN
early   = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
lr_sched = keras.callbacks.ReduceLROnPlateau(patience=4, factor=0.5, min_lr=1e-5)

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=300,
    batch_size=128,
    callbacks=[early, lr_sched],
    verbose=1
)

# 4)  EVALUATE
rmse = lambda yt, yp: np.sqrt(mean_squared_error(yt, yp))
val_pred  = model.predict(X_val,  verbose=0).ravel()
test_pred = model.predict(X_test, verbose=0).ravel()

print("\n=== Validation (2019) ===")
print(f"MAE  : {mean_absolute_error(y_val,  val_pred):.2f}")
print(f"RMSE : {rmse(y_val,  val_pred):.2f}")
print(f"R²   : {r2_score(y_val,  val_pred):.3f}")

print("\n=== Test (2020) ===")
print(f"MAE  : {mean_absolute_error(y_test, test_pred):.2f}")
print(f"RMSE : {rmse(y_test, test_pred):.2f}")
print(f"R²   : {r2_score(y_test, test_pred):.3f}")
```
